```python
import numpy as np; np.random.seed(0)
import matplotlib.pyplot as plt
import scipy.stats

a = dfm['sentiment']
b = dfp['sentiment']
c = dft['sentiment']

kdea = scipy.stats.gaussian_kde(a)
kdeb = scipy.stats.gaussian_kde(b)
kdec = scipy.stats.gaussian_kde(c)

both = np.hstack((a,b,c))
kdeboth = scipy.stats.gaussian_kde(both)
grid = np.arange(-1,1,0.00001)

#weighted kde curves
wa = kdea(grid)*(len(a)/float(len(both)))
wb = kdeb(grid)*(len(b)/float(len(both)))
wc = kdeb(grid)*(len(c)/float(len(both)))

print ("a.sum ", wa.sum())
print ("b.sum ", wb.sum())
print ("c.sum ", wc.sum())

fig, ax = plt.subplots()
ax.plot(grid, wa, lw=1, label = "McDonald's")
ax.plot(grid, wb, lw=1, label = "Panda Express")
ax.plot(grid, wc, lw=1, label = "Taco Bell")


plt.legend()
plt.show()
```
