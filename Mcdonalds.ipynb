```python
fetch_tweets = tweepy.Cursor(api.search, q = 'mcdonalds').items(10)
with open('mc.csv', mode = 'w', newline='',encoding='UTF-8') as my_file:
    my_writer = csv.writer(my_file,delimiter=',')
    my_writer.writerow(['created_at','text','hashtags','symbols','source',
                        'reply_screen_name','user_id','user_screen_name',
                       'user_location','user_description','followers_count',
                       'friends_count','listed_count','user_created_at',
                       'verified','user_lang','is_translator','geo',
                        'coordinates','place','contributes','is_quote',
                        'retweet_count','favorite_count','favorited','retweeted','lang'])
    for i in fetch_tweets:
        my_writer.writerow([i.created_at,i.text,i.entities['hashtags'],
                           i.entities['symbols'],
                           i.source,i.in_reply_to_screen_name,
                           i.user.id,i.user.screen_name,
                           i.user.location,i.user.description,
                           i.user.followers_count,i.user.friends_count,
                           i.user.listed_count,i.user.created_at,i.user.verified,
                           i.user.lang,i.user.is_translator,
                           i.geo,i.coordinates,i.place,i.contributors,
                           i.is_quote_status,i.retweet_count,i.favorite_count,
                           i.favorited,i.retweeted,i.lang])
```
